```python
plt.figure(figsize=(8,5))
df['Course_Name'].value_counts().plot(kind='bar')
plt.title("Course-wise Enrollment Count")
plt.xlabel("Course Name")
plt.ylabel("Number of Students")
plt.show()
```

```python
df['Enrollment_Date'] = pd.to_datetime(df['Enrollment_Date'])
df.groupby(df['Enrollment_Date'].dt.year)['Student_ID'].count().plot()
plt.title("Enrollment Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Enrollments")
plt.show()
```

```python
df['Placement_Status'].value_counts().plot(kind='pie', autopct='%1f%%')
plt.title("Placement Status Distribution")
plt.ylabel("")
plt.show()
```
